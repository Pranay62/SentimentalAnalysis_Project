```python
#import libraries
import pandas as pd
```

```python
df = pd.read_csv('AMAZON_MOBILE_DATA.csv')
# Read the csv using pandas
```

```python
df.head()
```

```python
#import libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
Sentiment = df['Rating'].value_counts()
plt.figure(figsize=(7, 6))
ax = Sentiment.plot(kind='bar', rot=0, color=sns.color_palette("RdYlGn", 6))
ax.set_title("Distribution of Ratings across the entire dataset", y = 1)
ax.set_xlabel('Ratings of Reviews')
ax.set_ylabel('Number of reviews corresponding to each of 5 ratings')
ax.set_xticklabels(('5', '1', '4', '3', '2'))

for rect in ax.patches:
    y_value = rect.get_height()
    x_value = rect.get_x() + rect.get_width() / 2
    space = 1
    label = "{:.0f}".format(y_value)
    ax.annotate(label, (x_value, y_value), xytext=(0, space), textcoords="offset points", ha='center', va='bottom')
plt.show()
```

```python
plt.figure(figsize=(10,6))
colors =['lightgreen', 'yellow', 'orange','blue','red']
explode=(0.1,0,0,0,0)
plt.pie(df['Rating'].value_counts().values, 
        explode=explode, 
        labels=df['Rating'].value_counts().index, 
        colors=colors, autopct='%1.0f%%', shadow=True)
plt.title('Ratings of Reviews')
plt.axis('equal')
plt.show()
```

```python
df['ReviewsLength']=df['Reviews'].str.len()
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = df['ReviewsLength']
hist, edges = np.histogram(x,bins=np.arange(0,2000,100))
plt.figure(figsize=(8, 6), dpi=80)
plt.bar(edges[:-1], hist, align="edge", ec="k", width=np.diff(edges))
plt.title('REVIEW LENTH DISTRIBUTION')
plt.xlabel('Review Length (Number of characters)')
plt.ylabel('Count')
plt.show()
```

```python
data3 = df[['Price','Reviews']]
#display(data3)
data3a=data3.groupby('Price')['Reviews'].count()
data3a = data3a.to_frame()
data3a
```

```python
plt.figure(figsize=(5,3), dpi=150)
plt.plot(data3a, marker='.', ms = 10,mec = 'blue',c='w', lw = '1')
plt.xticks(fontsize=10)
plt.xlabel('Price')
plt.ylabel('Count')
plt.title('PRICE Vs NUMBER OF REVIEWS')
plt.show()
```
